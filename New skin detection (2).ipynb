```python
# Step 2: Import YOLOv8
from ultralytics import YOLO
import time

# Step 3: Configure Dataset Paths
data_yaml_path = "C:/Users/ASUS/Music/DEEP LEARNING/New folder (2)/data.yaml"  # Path to the corrected data.yaml file

# Step 4: Initialize the Model with Your Trained Weights
model_path = "C:/Users/ASUS/Music/DEEP LEARNING/New folder (2)/runs/train/weights/best.pt"  # Path to the trained model weights
model = YOLO(model_path)

# Check if model loaded successfully
print("Model loaded successfully")

# Step 5: Run validation on the test dataset
metrics = model.val(
    data=data_yaml_path,
    split="test",  # Specify 'test' split for evaluation
    batch=2  # Use 'batch' instead of 'batch_size'
)

# Extract and display specific metrics
mAP = metrics["metrics/mAP_50"]           # Mean Average Precision at IoU=0.5
precision = metrics["metrics/precision"]  # Precision
recall = metrics["metrics/recall"]        # Recall
f1_score = 2 * (precision * recall) / (precision + recall)  # F1 Score

print(f"Mean Average Precision (mAP@0.5): {mAP:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")

# Step 6: Calculate Inference Time (On a Small Test Set)
test_image_path = "C:/Users/ASUS/Music/DEEP LEARNING/New folder (2)/test/"  # Path to the test images folder

start_time = time.time()
results = model.predict(source=test_image_path, save=False)
end_time = time.time()

inference_time = end_time - start_time
print(f"Inference Time: {inference_time:.4f} seconds")

# Optional: Display prediction results
results.show()
```
